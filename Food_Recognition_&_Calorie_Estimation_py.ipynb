```python
# =========================
# Task-05: Food Recognition & Calorie Estimation
# =========================

# Step 0: Install Required Libraries (Colab Only)
# !pip install tensorflow opencv-python matplotlib numpy pandas scikit-learn

import os
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.applications import MobileNetV2

# =========================
# Step 1: Create Dummy Dataset
# =========================
os.makedirs("dataset/food_images", exist_ok=True)

food_classes = {
    "Pizza": 266,
    "Burger": 354,
    "Salad": 152
}

data = []
for food, cal in food_classes.items():
    for i in range(5):  # 5 images per class
        img_name = f"{food.lower()}{i}.jpg"
        img_path = f"dataset/food_images/{img_name}"

        # Create random dummy image (128x128)
        img = np.random.randint(0, 255, (128, 128, 3), dtype=np.uint8)
        cv2.imwrite(img_path, img)

        # Slight variation in calories
        data.append([img_name, food, cal + np.random.randint(-10, 10)])

# Create CSV
df = pd.DataFrame(data, columns=["image", "label", "calories"])
df.to_csv("dataset/food_calories.csv", index=False)
print("✅ Dummy dataset created!")
print(df.head())

# =========================
# Step 2: Load Dataset
# =========================
image_folder = "dataset/food_images"
labels_csv = "dataset/food_calories.csv"

# Load CSV
data = pd.read_csv(labels_csv)
print("✅ CSV Loaded:", data.head())

# Parameters
img_size = (128, 128)

# Load images and labels
images, labels, calories = [], [], []
for i, row in data.iterrows():
    img_path = os.path.join(image_folder, row['image'])
    img = cv2.imread(img_path)
    img = cv2.resize(img, img_size)
    img = img / 255.0  # Normalize
    images.append(img)
    labels.append(row['label'])
    calories.append(row['calories'])

images = np.array(images)
calories = np.array(calories)
print("✅ Images Loaded:", images.shape)

# =========================
# Step 3: Encode Labels and Split Data
# =========================
label_encoder = LabelEncoder()
y_labels = label_encoder.fit_transform(labels)
y_labels_cat = to_categorical(y_labels)  # For classification

X_train, X_test, y_train_class, y_test_class, y_train_cal, y_test_cal = train_test_split(
    images, y_labels_cat, calories, test_size=0.2, random_state=42
)

# =========================
# Step 4: Build Multi-Task CNN Model
# =========================
base_model = MobileNetV2(weights=None, include_top=False, input_shape=(128,128,3))

x = Flatten()(base_model.output)

# Classification Head
class_output = Dense(len(label_encoder.classes_), activation='softmax', name='class_output')(x)

# Regression Head for Calorie Estimation
calorie_output = Dense(1, activation='linear', name='calorie_output')(x)

# Multi-output Model
model = Model(inputs=base_model.input, outputs=[class_output, calorie_output])

model.compile(
    optimizer='adam',
    loss={'class_output': 'categorical_crossentropy', 'calorie_output': 'mse'},
    metrics={'class_output': 'accuracy', 'calorie_output': 'mse'}
)

model.summary()

# =========================
# Step 5: Train the Model
# =========================
history = model.fit(
    X_train,
    {'class_output': y_train_class, 'calorie_output': y_train_cal},
    epochs=5,
    batch_size=4,
    validation_split=0.2
)

# =========================
# Step 6: Evaluate and Predict
# =========================
loss, class_loss, calorie_loss, class_acc, calorie_mse = model.evaluate(
    X_test,
    {'class_output': y_test_class, 'calorie_output': y_test_cal}
)
print(f"🎯 Classification Accuracy: {class_acc*100:.2f}%")
print(f"📉 Calorie MSE: {calorie_mse:.2f}")

# Predict on 5 test samples
y_pred_class, y_pred_cal = model.predict(X_test[:5])

for i in range(5):
    plt.imshow(X_test[i])
    plt.axis('off')
    plt.title(
        f"Pred: {label_encoder.inverse_transform([np.argmax(y_pred_class[i])])[0]}\n"
        f"Calories: {y_pred_cal[i][0]:.1f}, Actual: {y_test_cal[i]}"
    )
    plt.show()
```
